```python
import numpy as np
import pandas as pd
import ast
```

```python
movies = pd.read_csv('datasets/bollywood.csv')
movies
```

```python
movies = movies[['Movie_Name', 'Genre','Lead_Star','Director','Music_Director']]
movies
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace = True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].Genre
```

```python
def to_list(val):
    try:
        return ast.literal_eval(val)
    except (ValueError, SyntaxError):
        return [val]  # just wrap it in a list if eval fails
```

```python
movies['Genre'] = movies['Genre'].apply(to_list)
```

```python
movies.head()
```

```python
movies['Lead_Star'] = movies['Lead_Star'].apply(to_list)
movies['Director'] = movies['Director'].apply(to_list)
movies['Music_Director'] = movies['Music_Director'].apply(to_list)
```

```python
movies.head()
```

```python
movies['Genre'] = movies['Genre'].apply(lambda x:[i.replace(" ","_") for i in x])
movies['Lead_Star'] = movies['Lead_Star'].apply(lambda x:[i.replace(" ","_") for i in x])
movies['Director'] = movies['Director'].apply(lambda x:[i.replace(" ","_") for i in x])
movies['Music_Director'] = movies['Music_Director'].apply(lambda x:[i.replace(" ","_") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['Genre'] + movies['Lead_Star'] + movies['Director'] + movies['Music_Director']
```

```python
movies.head()
```

```python
df = movies[['Movie_Name','tags']]
df
```

```python
df["tags"] = df['tags'].apply(lambda x:" ".join(x))
```

```python
df["tags"] = df["tags"].apply(lambda x:x.lower())
```

```python
df.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
vectors = cv.fit_transform(df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
similarity
```

```python
df["Movie_Name"] = df["Movie_Name"].apply(lambda x:x.lower())
```

```python
df
```

```python
def recommend(movie):
    # Check if movie exists in the DataFrame
    if movie not in df['Movie_Name'].values:
        print(f"'{movie}' not found in dataset.")
        return

    # Get index using reset index logic
    movie_index = df[df['Movie_Name'] == movie].index
    movie_index = movie_index[0]

    # Fix: convert movie_index to positional index (0-based) if new_df index is not 0-based
    movie_index = df.reset_index().query("Movie_Name == @movie").index[0]

    # Now safe to use in similarity
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]

    for i in movie_list:
        print(df.iloc[i[0]].Movie_Name)
```

```python
recommend('mr. hot mr. kool')
```

```python
import joblib

# Save the necessary components
joblib.dump(df, 'movies_df.joblib',compress=5)
joblib.dump(similarity, 'similarity_matrix.joblib',compress=5)
joblib.dump(cv, 'count_vectorizer.joblib',compress=5)
```
