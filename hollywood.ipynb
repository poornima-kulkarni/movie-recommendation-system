```python
import numpy as np
import pandas as pd
import ast
```

```python
movie = pd.read_csv("datasets/movies.csv")
movie
```

```python
credit = pd.read_csv("datasets/credits.csv")
credit
```

```python
movie = movie.merge(credit,on="title")
movie
```

```python
movie.tail(1)
```

```python
movie = movie[["keywords","title","movie_id","cast","crew","overview","genres"]]
movie
```

```python
movie.isnull().sum()
```

```python
movie.dropna(inplace = True)
```

```python
movie.duplicated().sum()
```

```python
movie.iloc[0].genres
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
    
```

```python
movie['genres'] = movie['genres'].apply(convert)
```

```python
movie.head()
```

```python
movie['keywords'] = movie['keywords'].apply(convert)
```

```python
movie.head()
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L
```

```python
movie['cast'] = movie['cast'].apply(convert3)
```

```python
movie.head()
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movie['crew'] = movie['crew'].apply(fetch_director)
```

```python
movie.head()
```

```python
movie['overview'] = movie['overview'].apply(lambda x:x.split())
```

```python
movie.head()
```

```python
movie['genres'] = movie['genres'].apply(lambda x:[i.replace(" ","_") for i in x])
movie['keywords'] = movie['keywords'].apply(lambda x:[i.replace(" ","_") for i in x])
movie['cast'] = movie['cast'].apply(lambda x:[i.replace(" ","_") for i in x])
movie['crew'] = movie['crew'].apply(lambda x:[i.replace(" ","_") for i in x])
```

```python
movie.head()
```

```python
movie['tags'] = movie['genres'] + movie['keywords'] + movie['overview'] + movie['cast'] + movie['crew']
```

```python
movie.head()
```

```python
new_df = movie[['movie_id','title','tags']]
new_df
```

```python
new_df["tags"] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df["tags"] = new_df["tags"].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
similarity
```

```python
new_df
```

```python
def recommend(movie):
    # Check if movie exists in the DataFrame
    if movie not in new_df['title'].values:
        print(f"'{movie}' not found in dataset.")
        return

    # Get index using reset index logic
    movie_index = new_df[new_df['title'] == movie].index
    movie_index = movie_index[0]

    # Fix: convert movie_index to positional index (0-based) if new_df index is not 0-based
    movie_index = new_df.reset_index().query("title == @movie").index[0]

    # Now safe to use in similarity
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]

    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('My Date with Drew')
```

```python
import joblib


# Save necessary components
joblib.dump(new_df, 'new_df.joblib')
joblib.dump(similarity, 'similarity_hollywood.joblib',compress=3)
joblib.dump(cv, 'new_vectorizer.joblib')
```
